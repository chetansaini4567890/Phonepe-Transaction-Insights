## Inspect Aggregated Transaction Path

```python
agg_txn_path = os.path.join(RAW_DATA_PATH, "aggregated", "transaction", "country", "india", "state")
print("Aggregated Transaction Path Exists:", os.path.exists(agg_txn_path))
```

```python
if os.path.exists(agg_txn_path):
    states = os.listdir(agg_txn_path)
    print("Number of states found:", len(states))
    print("Sample states:", states[:10])
```

## Open One Sample JSON File

```python
sample_state = states[0]
sample_year_path = os.path.join(agg_txn_path, sample_state, os.listdir(os.path.join(agg_txn_path, sample_state))[0])
sample_file = os.listdir(sample_year_path)[0]
sample_file_path = os.path.join(sample_year_path, sample_file)

print("Sample file path:")
print(sample_file_path)
```

```python
with open(sample_file_path, "r") as f:
    sample_json = json.load(f)

type(sample_json)
```
