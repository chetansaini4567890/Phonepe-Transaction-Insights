# Notebook 7: SQL Business Case Analysis

## Objective
In this notebook, we will use SQL queries to answer important business questions related to transactions, users, insurance, geography, and payment behavior.

## Why this notebook matters
The project specifically requires SQL-based business case analysis before Python visualization and dashboarding.

```python
import pandas as pd
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings("ignore")
```

```python
DB_USER = "root"
DB_PASSWORD = "12345"
DB_HOST = "localhost"
DB_PORT = "3306"
DB_NAME = "phonepe_insights"

engine = create_engine(
    f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)
```

## Business Case 1: Top 10 States by Transaction Amount

```python
query_1 = """
SELECT state, ROUND(SUM(transaction_amount), 2) AS total_transaction_amount
FROM aggregated_transaction
GROUP BY state
ORDER BY total_transaction_amount DESC
LIMIT 10;
"""
bc1 = pd.read_sql(query_1, engine)
bc1
```

### Insight
This query identifies the states contributing the highest total transaction amount. These states can be treated as priority markets for premium services, retention campaigns, and deeper product expansion.

## Business Case 2: Top 10 States by Transaction Count

```python
query_2 = """
SELECT state, SUM(transaction_count) AS total_transaction_count
FROM aggregated_transaction
GROUP BY state
ORDER BY total_transaction_count DESC
LIMIT 10;
"""
bc2 = pd.read_sql(query_2, engine)
bc2
```

### Insight
This analysis highlights transaction activity volume rather than monetary value, helping distinguish high-frequency usage regions from high-ticket-value regions.

## Business Case 3: Category-wise Transaction Performance

```python
query_3 = """
SELECT transaction_type, 
       SUM(transaction_count) AS total_count,
       ROUND(SUM(transaction_amount), 2) AS total_amount
FROM aggregated_transaction
GROUP BY transaction_type
ORDER BY total_amount DESC;
"""
bc3 = pd.read_sql(query_3, engine)
bc3
```

### Insight
This query helps identify which payment categories dominate the platform. It is useful for payment optimization, feature prioritization, and business strategy.

## Business Case 4: Year-wise Transaction Trend

```python
query_4 = """
SELECT year,
       ROUND(SUM(transaction_amount), 2) AS total_amount,
       SUM(transaction_count) AS total_count
FROM aggregated_transaction
GROUP BY year
ORDER BY year;
"""
bc4 = pd.read_sql(query_4, engine)
bc4
```

### Insight
This trend analysis shows overall growth or decline across years and helps assess platform expansion over time.

## Business Case 5: Quarter-wise Transaction Trend

```python
query_5 = """
SELECT quarter,
       ROUND(SUM(transaction_amount), 2) AS total_amount,
       SUM(transaction_count) AS total_count
FROM aggregated_transaction
GROUP BY quarter
ORDER BY quarter;
"""
bc5 = pd.read_sql(query_5, engine)
bc5
```

### Insight
Quarter-level analysis reveals seasonal or cyclical transaction behavior that may be useful for campaign planning and resource allocation.

## Business Case 6: Top Districts by Transaction Amount

```python
query_6 = """
SELECT district,
       state,
       ROUND(SUM(transaction_amount), 2) AS total_amount
FROM map_transaction
GROUP BY district, state
ORDER BY total_amount DESC
LIMIT 10;
"""
bc6 = pd.read_sql(query_6, engine)
bc6
```

### Insight
This query identifies local high-value regions for district-level targeting and strategic regional prioritization.

## Business Case 7: Top Pincode or Entity Performance

```python
query_7 = """
SELECT entity_name,
       entity_type,
       state,
       ROUND(SUM(transaction_amount), 2) AS total_amount
FROM top_transaction
GROUP BY entity_name, entity_type, state
ORDER BY total_amount DESC
LIMIT 10;
"""
bc7 = pd.read_sql(query_7, engine)
bc7
```

### Insight
This helps locate top-performing granular entities such as districts or pincodes based on the transformed top-level dataset structure.

## Business Case 8: State-wise Registered Users

```python
query_8 = """
SELECT state,
       SUM(registered_users) AS total_registered_users
FROM map_user
GROUP BY state
ORDER BY total_registered_users DESC
LIMIT 10;
"""
bc8 = pd.read_sql(query_8, engine)
bc8
```

### Insight
Registered user concentration indicates platform reach and can be used for acquisition, onboarding, and retention planning.

## Business Case 9: Brand-wise User Share

```python
query_9 = """
SELECT brand,
       SUM(count) AS total_users,
       ROUND(AVG(percentage), 2) AS avg_percentage
FROM aggregated_user
GROUP BY brand
ORDER BY total_users DESC;
"""
bc9 = pd.read_sql(query_9, engine)
bc9
```

### Insight
Brand distribution helps understand device-level user presence and can support compatibility, UX, and partnership decisions.

## Business Case 10: Insurance Performance by State

```python
query_10 = """
SELECT state,
       ROUND(SUM(transaction_amount), 2) AS total_insurance_amount,
       SUM(transaction_count) AS total_insurance_count
FROM aggregated_insurance
GROUP BY state
ORDER BY total_insurance_amount DESC
LIMIT 10;
"""
bc10 = pd.read_sql(query_10, engine)
bc10
```

### Insight
This analysis highlights the strongest markets for insurance-related offerings and helps identify expansion opportunities.

## Business Case 11: Insurance Trend Over Time

```python
query_11 = """
SELECT year,
       ROUND(SUM(transaction_amount), 2) AS total_insurance_amount,
       SUM(transaction_count) AS total_insurance_count
FROM aggregated_insurance
GROUP BY year
ORDER BY year;
"""
bc11 = pd.read_sql(query_11, engine)
bc11
```

### Insight
A time-based insurance trend reveals whether platform users are adopting financial protection products more actively over time.

## Business Case 12: State-Year Growth Comparison

```python
query_12 = """
SELECT state, year,
       ROUND(SUM(transaction_amount), 2) AS total_amount
FROM aggregated_transaction
GROUP BY state, year
ORDER BY state, year;
"""
bc12 = pd.read_sql(query_12, engine)
bc12.head(20)
```

### Insight
This state-year comparison prepares the data for later trend visualization and helps identify high-growth and low-growth markets.
