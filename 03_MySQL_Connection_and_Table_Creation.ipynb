# PhonePe Transaction Insights
## Notebook 3: MySQL Connection and Table Creation

### Objective
This notebook connects to the MySQL database and creates the tables required to store the extracted PhonePe Pulse data.

### Tables Covered
- aggregated_transaction
- aggregated_user
- map_transaction
- map_user
- top_transaction
- top_user
- aggregated_insurance

## Important
Before running this notebook, create the database manually in MySQL:

```sql
CREATE DATABASE phonepe_insights;
```

````python
import pandas as pd
from sqlalchemy import create_engine, text
````

## Set Database Credentials
Replace the placeholder password with your MySQL password.

````python
MYSQL_USERNAME = "root"
MYSQL_PASSWORD = "12345"
MYSQL_HOST = "localhost"
MYSQL_DATABASE = "phonepe_insights"
````

````python
engine = create_engine(
    f"mysql+pymysql://{MYSQL_USERNAME}:{MYSQL_PASSWORD}@{MYSQL_HOST}/{MYSQL_DATABASE}"
)

print("SQLAlchemy engine created successfully.")
````

````python
with engine.connect() as connection:
    print("Connected to MySQL database successfully.")
````

## Create Required Tables

````python
create_aggregated_transaction = """
CREATE TABLE IF NOT EXISTS aggregated_transaction (
    id INT AUTO_INCREMENT PRIMARY KEY,
    state VARCHAR(100),
    year INT,
    quarter INT,
    transaction_type VARCHAR(100),
    transaction_count BIGINT,
    transaction_amount DOUBLE
);
"""
````

````python
create_aggregated_user = """
CREATE TABLE IF NOT EXISTS aggregated_user (
    id INT AUTO_INCREMENT PRIMARY KEY,
    state VARCHAR(100),
    year INT,
    quarter INT,
    brand VARCHAR(100),
    count BIGINT,
    percentage DOUBLE
);
"""
````

````python
create_map_transaction = """
CREATE TABLE IF NOT EXISTS map_transaction (
    id INT AUTO_INCREMENT PRIMARY KEY,
    state VARCHAR(100),
    year INT,
    quarter INT,
    district VARCHAR(150),
    transaction_count BIGINT,
    transaction_amount DOUBLE
);
"""
````

````python
create_map_user = """
CREATE TABLE IF NOT EXISTS map_user (
    id INT AUTO_INCREMENT PRIMARY KEY,
    state VARCHAR(100),
    year INT,
    quarter INT,
    district VARCHAR(150),
    registered_users BIGINT,
    app_opens BIGINT
);
"""
````

````python
create_top_transaction = """
CREATE TABLE IF NOT EXISTS top_transaction (
    id INT AUTO_INCREMENT PRIMARY KEY,
    state VARCHAR(100),
    year INT,
    quarter INT,
    entity_type VARCHAR(50),
    entity_name VARCHAR(150),
    transaction_count BIGINT,
    transaction_amount DOUBLE
);
"""
````

````python
create_top_user = """
CREATE TABLE IF NOT EXISTS top_user (
    id INT AUTO_INCREMENT PRIMARY KEY,
    state VARCHAR(100),
    year INT,
    quarter INT,
    entity_type VARCHAR(50),
    entity_name VARCHAR(150),
    registered_users BIGINT
);
"""
````

````python
create_aggregated_insurance = """
CREATE TABLE IF NOT EXISTS aggregated_insurance (
    id INT AUTO_INCREMENT PRIMARY KEY,
    state VARCHAR(100),
    year INT,
    quarter INT,
    insurance_type VARCHAR(100),
    transaction_count BIGINT,
    transaction_amount DOUBLE
);
"""
````

## Execute Table Creation Queries

````python
table_queries = [
    create_aggregated_transaction,
    create_aggregated_user,
    create_map_transaction,
    create_map_user,
    create_top_transaction,
    create_top_user,
    create_aggregated_insurance
]

with engine.connect() as connection:
    for query in table_queries:
        connection.execute(text(query))
    connection.commit()

print("All tables created successfully.")
````

## Verify Table Creation

````python
show_tables_query = "SHOW TABLES;"

with engine.connect() as connection:
    result = connection.execute(text(show_tables_query))
    tables = result.fetchall()

tables
````

````python
print("Tables available in database:")
for table in tables:
    print("-", table[0])
````
